```python
rating = df_merge.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

Observation


1.   When rating is between  4.5 to 4.9 ------> Excellent

2.  when rating is between 4.0 to 4.4 ------> very good
3.  when rating is between 3.0 to 3.5 -----> good
4.   when rating is between 2.5 to 2.9 ----> average
5.  when rating is between bleow to 2.5 ----> poor
6.  most of people is not given rating






```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,8)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=rating,palette=['grey','red','orange','yellow','green','green'])
```

Observation :

1.Not rated count is high

2.maximum number of rating between 2.5 to 3.9

```python
sns.countplot(x='Rating color',data = rating,palette=['grey','red','orange','yellow','green','green'])
```
